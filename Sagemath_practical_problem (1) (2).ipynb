**Q.26 	Consider the function**  $f(x)=e^{((x-1)^2 )}+sin⁡(cos⁡(5x))$**. Verify the Lagrange’s Mean Value Theorem for f in the interval \[1,2\]. Plot the graph to explain the geometric meaning.**

**Solution:**


```python
f(x) = exp((x - 1)^2) + sin(cos(5*x))
a, b = 1, 2
m = (f(b) - f(a)) / (b - a)
show(m)
```

```python
f1 = f.diff()
show(f1)
```

```python
h(x) = f1 - m
c1 = h.find_root(a,b)
show(c1)
```

```python
c2 = h.find_root(a,1.5)
show(c2)
```

```python
t1(x) = f(c1) + (f1(c1) * (x - c1))
show(t1(x))
```

```python
t2(x) = f(c2) + (f1(c2) * (x - c2))
show(t2(x))
```

```python
fp = plot(f(x), (a,b), color = 'red')
show(fp)
```

```python
chord = line([(a,f(a)), (b,f(b))], color = 'green')
show(chord)
```

```python
tg1 = plot(t1(x), (c1-0.5, c1+0.5))
show(tg1)
```

```python
tg2 = plot(t2, (c2-0.5, c2+0.5))
show(tg2)
```

```python
show(fp + chord + tg1 + tg2, figsize = 5)
```

**Q.27 Consider f\(x\) = e^{−x^2}cos 2x.**

**\(i\) Plot the graph of the function along with Taylor’s polynomial of degree 1, 2,3, 6,7, 9, 10.**

**\(ii\) Tabulate error term of the approximation of f\(x\) at x = 1.2 when it is approximated by Taylor’s polynomial of degree 1 to 10 about x = 1.**


**\(i\) Plot the graph of the function along with Taylor’s polynomial of degree 1, 2,3, 6,7, 9, 10.**


```python
f(x) = exp(-x^2) * cos(2*x)
p = plot(f(x), (-3,1), color = 'red')
plot(p)
```

```python
col = ['yellow', 'black', 'green', 'brown', 'blue', 'grey', 'orange']
deg = [1, 2,3, 6,7, 9, 10]
for i in range(len(deg)):
    t = f.taylor(x, 1, i)
    show(t)
    p = p + plot(t(x), (-3, 1), color = col[i])
    show(p, figsize = 5, ymin=-0.5, ymax=2)
```

**\(ii\) Tabulate error term of the approximation of f\(x\) at x = 1.2 when it is approximated by Taylor’s polynomial of degree 1 to 10 about x = 1.**

**Solution :The errors can be tabulated with the following code.**


```python
f(x) = exp(-x^2) * cos(2*x)
a = 1
x1 = 1.2
error = []
T=[]
print('Actual value of the function at f(1.2)={}'.format(f(x=x1)))
for i in range(1,11):
    tn = f.taylor(x, 1, i)
    error.append(abs(tn(x1).n() - f(x1).n()))
    T.append(tn(x=x1).n())
table([(i,T[i-1],error[i-1]) for i in [1,2,..,10]], header_row= ["n", 'tn(x=1.2)',r"Error term"], frame = True)
```

**Q,28 	Explain Geometric meaning of Cauchy mean value theorem for f\(t\)= t cos t and g\(t\)= t sin t  for 1≤t≤10**

**olution:**


```python
var('t')
f(t) = t*cos(t)
g(t) = t*sin(t)
a,b = 1, 10
curve = parametric_plot((f(t),g(t)), (t,a,b))
show(curve)
```

```python
end_pts = points([(f(a),g(a)),(f(b),g(b))],size=20,color='green')
show(end_pts)
```

```python
chrd = line([(f(a),g(a)),(f(b),g(b))],color='red',linestyle=':')
show(chrd)
```

```python
show(curve+chrd+end_pts,figsize=4)
```

```python
f1 = f.diff(t)
g1 = g.diff(t)
m = (f(b)-f(a))/(g(b)-g(a))
show(m.n())
h(t) = f1(t)/g1(t)-m
h1=plot(h(t),a,b,ymax=5,ymin=-2,figsize=4)
show(h1)
```

```python
c1 = h.find_root(2,4)
show(c1)
c2 = h.find_root(5,6)
show(c2)
c3 = h.find_root(8,10)
show(c3)
```

```python
t1 = g(c1)+g1(c1)/f1(c1)*(x-f(c1))
show(t1)
```

```python
tgt1 = plot(t1,f(c1)-2,f(c1)+2,color='green')
show(tgt1)
```

```python
p1 = point((f(c1),g(c1)),size=20,color='green')
show(p1)
```

```python
t2 = g(c2)+g1(c2)/f1(c2)*(x-f(c2))
show(t2)
```

```python
tgt2 = plot(t2,f(c2)-2,f(c2)+2,color='black')
show(tgt2)
```

```python
p2 = point((f(c2),g(c2)),size=20,color='black')
show(p2)
```
